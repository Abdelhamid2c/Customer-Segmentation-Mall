### Percentile Method

```python
data.Gender.value_counts(normalize= True)
```

```python
print(f'max age : {data.Age.max()}')
print(f'min age : {data.Age.min()}')
```

```python
p1, p2 = np.percentile(data.Age, [33.3, 66.6])
p1
```

```python
data['Age Bin'] = np.where(data.Age > p2, 'High', 
                           np.where(data.Age > p1, 'Medium' ,
                                    'Low')
                          )
data['Age Bin'].value_counts()
```

```python
def binner_fct(column, data):
    p1, p2 = np.percentile(data.Age, [33.3, 66.6])
    data[f'{column} Bin'] = np.where(data[column] > p2, 'High', np.where(data[column] > p1, 'Medium' , 'Low'))
```

```python
columns = ['Annual Income (k$)', 'Spending Score (1-100)']

for col in columns:
    binner_fct(col,data)
```

```python
data.groupby(['Gender', 
              'Age Bin', 
              'Annual Income (k$) Bin', 
              'Spending Score (1-100) Bin']).mean()[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
data.groupby(['Gender', 
              'Age Bin', 
              'Annual Income (k$) Bin', 
              'Spending Score (1-100) Bin'])\
            .count()[['CustomerID']]\
            .sort_values(by = 'CustomerID', ascending=False)
```

### KMeans

```python
num_clusters = 3

X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
kmeans = KMeans(num_clusters, random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

```python
data['Cluster name'] = labels
```

### PCA

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['Age'], X['Annual Income (k$)'], X['Spending Score (1-100)'], c=kmeans.labels_)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
