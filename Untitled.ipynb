### Percentile Method

```python
data.Gender.value_counts(normalize= True)
```

```python
print(f'max age : {data.Age.max()}')
print(f'min age : {data.Age.min()}')
```

```python
p1, p2 = np.percentile(data.Age, [33.3, 66.6])
p1
```

```python
data['Age Bin'] = np.where(data.Age > p2, 'High', 
                           np.where(data.Age > p1, 'Medium' ,
                                    'Low')
                          )
data['Age Bin'].value_counts()
```

```python
def binner_fct(column, data):
    p1, p2 = np.percentile(data.Age, [33.3, 66.6])
    data[f'{column} Bin'] = np.where(data[column] > p2, 'High', np.where(data[column] > p1, 'Medium' , 'Low'))
```

```python
columns = ['Annual Income (k$)', 'Spending Score (1-100)']

for col in columns:
    binner_fct(col,data)
```

### KMeans

```python
num_clusters = 3

X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
kmeans = KMeans(num_clusters, random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

```python
data['Cluster name'] = labels
```

### how to determine the numbers of clusters

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np


# Run k-means clustering for k=2 to 10 clusters
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++',
                    max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Plot the silhouette scores
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Analysis')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
